Demonstrates three ADK concepts: tools (Google Custom Search), memory persistence, and artifact (PDF export).

```python
import os
os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY', 'dummy')
os.environ['GOOGLE_CSE_ID'] = os.getenv('GOOGLE_CSE_ID', '')
from app.agents.dream_interpreter import DreamInterpreterAgent
from app.agents.research_agent import ResearchAgent
from app.agents.logistics_agent import LogisticsAgent
from app.agents.budget_agent import BudgetAgent
from app.agents.memory_agent import MemoryAgent
from app.models.trip import TripPlan

dream = '5-day Tokyo trip focused on culture and food, moderate budget'
di = DreamInterpreterAgent(api_key=os.environ['GOOGLE_API_KEY'])
params = await di.process(dream) if hasattr(di, 'process') else None
ra = ResearchAgent(api_key=os.environ['GOOGLE_API_KEY'], google_api_key=os.environ['GOOGLE_API_KEY'], google_cse_id=os.environ['GOOGLE_CSE_ID'])
findings = await ra.process(params) if hasattr(ra, 'process') else None
la = LogisticsAgent(api_key=os.environ['GOOGLE_API_KEY'])
itinerary = await la.process(params, findings) if hasattr(la, 'process') else []
ba = BudgetAgent(api_key=os.environ['GOOGLE_API_KEY'])
budget = await ba.process(params, itinerary) if hasattr(ba, 'process') else {}
plan = TripPlan(parameters=params, itinerary=itinerary, budget_info=budget, status='generated')
plan
```
